```python
data = data.drop('Clock_in', 1)
data = data.drop('Clock_in_date', 1)
data = data.drop('Clock_in_time', 1)

data = data.drop('Clock_out', 1)
data = data.drop('Clock_out_time', 1)
data = data.drop('Clock_out_date', 1)
```

```python
data.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
data['RFID'] = labelencoder.fit_transform(data['RFID'])
data['time_interval'] = labelencoder.fit_transform(data['time_interval'])
data['real_interval'] = labelencoder.fit_transform(data['real_interval'])
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop('target', 1)
y= data["target"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2018)
x_train, x_val, y_train, y_val = train_test_split(x_train,y_train,test_size=0.1,random_state=2018)
#Check the sizes of all newly created datasets
print("Shape of x_train:",x_train.shape)
print("Shape of x_val:",x_val.shape)
print("Shape of x_test:",x_test.shape)
print("Shape of y_train:",y_train.shape)
print("Shape of y_val:",y_val.shape)
print("Shape of y_test:",y_test.shape)
```

### Defining Model Baseline Performance

```python
#calculate the average score of the train dataset
mean_target= y_train.mean()
print("baseline mean :",mean_target)
```

### Designing the DNN Model

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(256,input_dim = x_train.shape[1],activation="relu"))
model.add(Dense(256,activation="relu"))

model.add(Dense(256,activation="relu"))

model.add(Dense(256,activation="relu"))

model.add(Dense(256,activation="relu"))

model.add(Dense(256,activation="relu"))

model.add(Dense(1,activation = "sigmoid")) 

#activation =sigmoid for binary classification

model.compile(optimizer = "Adam",loss="binary_crossentropy",metrics=["accuracy"])

model.fit(x_train,y_train, validation_data = (x_val,y_val),epochs=10, batch_size=20)
```

```python
result = model.evaluate(x_test,y_test)
for i in range(len(model.metrics_names)):
    print("Metric ",model.metrics_names[i],":",str(round(result[i],2)))
```
